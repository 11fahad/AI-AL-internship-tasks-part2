```python
# Task 5 - Auto Tagging Support Tickets (Pure Python)

# -----------------------
# 1. Define Categories & Keywords
# -----------------------
categories = {
    "billing": ["invoice", "payment", "charged", "bill", "refund"],
    "technical_issue": ["error", "crash", "bug", "issue", "not working"],
    "account": ["login", "password", "account", "username", "locked"],
    "feature_request": ["feature", "add", "new", "improve", "request"],
    "bug": ["crash", "freeze", "broken", "fix", "bug"]
}

# -----------------------
# 2. Simple Scoring Function
# -----------------------
def tag_ticket(text):
    text_lower = text.lower()
    scores = {}
    for category, keywords in categories.items():
        score = 0
        for kw in keywords:
            if kw in text_lower:
                score += 1
        scores[category] = score
    # Sort categories by score
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return sorted_scores[:3]  # top 3 tags

# -----------------------
# 3. Test Dataset
# -----------------------
tickets = [
    "My invoice shows I was charged twice this month",
    "The app crashes when I upload a photo",
    "I can't reset my password, my account is locked",
    "Please add a dark mode feature",
    "The screen freezes when I try to open settings"
]

# -----------------------
# 4. Run Auto-Tagging
# -----------------------
for t in tickets:
    print(f"\nTicket: {t}")
    print("Top Tags:", tag_ticket(t))

categories = {
    "billing": ["invoice", "payment", "charged", "bill", "refund"],
    "technical_issue": ["error", "crash", "bug", "issue", "not working"],
    "account": ["login", "password", "account", "username", "locked"],
    "feature_request": ["feature", "add", "new", "improve", "request"],
    "bug": ["crash", "freeze", "broken", "fix", "bug"]
}

# -----------------------
# 2. Simple Scoring Function
# -----------------------
def tag_ticket(text):
    text_lower = text.lower()
    scores = {}
    for category, keywords in categories.items():
        score = 0
        for kw in keywords:
            if kw in text_lower:
                score += 1
        scores[category] = score
    # Sort categories by score
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return sorted_scores[:3]  # top 3 tags

# -----------------------
# 3. Test Dataset
# -----------------------
tickets = [
    "My invoice shows I was charged twice this month",
    "The app crashes when I upload a photo",
    "I can't reset my password, my account is locked",
    "Please add a dark mode feature",
    "The screen freezes when I try to open settings"
]

# -----------------------
# 4. Run Auto-Tagging
# -----------------------
for t in tickets:
    print(f"\nTicket: {t}")
    print("Top Tags:", tag_ticket(t))
```
